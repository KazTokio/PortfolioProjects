# Text Adventure Game

## Description

Welcome to "A Game with Jigsaw”.A bone-chilling, nerve-racking text-based adventure game that's gonna get your heart pumping. It's a thrilling experience inspired by the "Saw" movie series. In the creepy world of Saw, survival is a game, and the mysterious "Jigsaw" serial killer is playing to win. It's an adventurous challenge that you have to go through from the comfort of your own home. You're stuck in a dark, spooky, series of rooms, each filled with trials. Are you ready to test your resourcefulness, courage, and decision-making skills? Let's play the game.


### Story

You awake in a dimly lit room. You can't see anything clearly and you have no clue how or why you're here. A screen flickers on, illuminating the dark room. There on the screen is the terrifying visage of the Jigsaw Killer himself. He tells you that you are a part of his twisted game of survival. To live, you have to make your way through various stages, each filled with mind-boggling puzzles and life-altering decisions. Remember, every choice you make carries massive weight. The dice is rolled, the game has commenced, and survival is your only gambit.

### Objective

- Complete each stage by solving puzzles and making the right choices.
- Survive and escape from the hands of the Jigsaw Killer.

### Contents

- **Stage 1**: Find the keys hidden in the body parts to disarm the bombs.
- **Stage 2**: Continue the sequence of numbers displayed on the screen.
- **Stage 3**: Unscramble the sentence to escape from the room.
